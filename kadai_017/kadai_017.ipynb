```python
class Human:
    def __init__(self, name, age):

        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name}は大人です。")
        else:
            print(f"{self.name}は未成年です。")

human_list = []

human_list.append(Human("侍太郎", 30))
human_list.append(Human("侍花子", 20))
human_list.append(Human("侍次郎", 10))

for human in human_list:
    human.check_adult()
```
